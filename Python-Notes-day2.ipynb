# Python Set

```python
Image('3.png')
```

### 1. Set Operations
     A. Union 
     B. Intersection
     C. Difference
     D. Symmetric difference

#### 1.A) Union:

```python
A = {0, 2, 4, 6, 8,7,8,5,7,8,5,7}
B = {1, 2, 3, 4, 5,7,5,8}
print("Union :", A | B) 
```

#### 1.B. Intersection

```python
A = {0, 2, 4, 6, 8,7,8,5,7,8,5,7}
B = {1, 2, 3, 4, 5,7,5,8}
print("Intersection :", A & B) 
```

#### 1.C. Difference

```python
A = {0, 2, 4, 6, 8,7,8,5,7,8,5,7}
B = {1, 2, 3, 4, 5,7,5,8}
print("Difference :", A - B) 
```

####  1.D. Symmetric difference:

```python
A = {0, 2, 4, 6, 8,7,8,5,7,8,5,7}
B = {1, 2, 3, 4, 5,7,5,8}
print("Symmetric difference :", A ^ B) 
```

### 1. Built-in Dictionary functions

#### 1.A) len(dict):

```python
dict = {'Name': 'vema', 'Age': 27};
len (dict)
```

#### 2.B) str(dict):

```python
dict = {'Name': 'vema', 'Age': 27};
str (dict)
```

#### 1.C) type(variable):

```python
dict = {'Name': 'vema', 'Age': 27};
type(dict)
```

### 2.Built-in Dictionary methods

    A.	dic.clear()	  =It is used to delete all the items of the dictionary.
    B.	dict.copy()	    =It returns a shallow copy of the dictionary.
    C.	dict.fromkeys()	=Create a new dictionary from the iterable with the values equal to value.
    D.  dict.get(key[, value])    	=It is used to get the value specified for the passed key.
    F.	dict.items()	=It returns all the key-value pairs as a tuple.
    I.	dict.keys() 	=It returns all the keys of the dictionary.
    J.	dict.setdefault()	=It is used to set the key to the default value if the key is not specified in the dictionary
    K.	dict.update()	 =It updates the dictionary by adding the key-value pair of dict2 to this dictionary.
    L.	dict.values()	=It returns all the values of the dictionary.

#### 2.A) dic.clear():

```python
dictionaries = {1: "GREAT", 2: "LEARNING"}
dictionaries.clear()
dictionaries
```

#### 2.B) dict.copy():

```python
Dictionaries = {1: "GREAT", 2: "LEARNING"}
new = Dictionaries.copy()

print('Orignal: ', Dictionaries)
print('New: ', new)
```

#### 2.C) dict.fromkeys():

```python
keys = {'a', 'e', 'i', 'o', 'u' }
Dictionaries = dict.fromkeys(keys)
print(Dictionaries)
```

#### 2.D) dict.get(key[, value]) :

```python
Dictionaries = {'Name': 'vema', 'Age': 27};

print('Name: ', Dictionaries.get('Name'))
print('Age: ', Dictionaries.get('age'))
print('Salary: ', Dictionaries.get('salary'))
print('Salary: ', Dictionaries.get('salary', 0.0))
```

#### 2.E) dict.items()

```python
sales = { 'apple': 1, 'orange': 32, 'grapes': 42 }
print(sales.items())
```

#### 2.F) dict.keys() :

```python
Dictionaries = {'Name': 'vema', 'Age': 27};
print(Dictionaries.keys())
empty_dict = {}
print(empty_dict.keys())
```

#### 2.G) dict.setdefault():

```python
Dict = {'Name': 'vema', 'Age': 27};
print(Dictionaries.keys())
age = Dict.setdefault('age')
print('person = ',Dict)
print('Age = ',age)
```

#### 2.H) dict.update():

```python
Dict1 = {1: "one", 2: "three"}
Dict2 = {2: "two"}
Dict1.update(Dict1)
print(Dict1)

Dict2 = {3: "three"}
# adds element with key 3
Dict1.update(Dict2)
print(Dict1)
```

 #### 2.L)  dict.values():

```python
Dict = { 'apple': 2, 'orange': 3, 'grapes': 4 }
print(Dict.values())
```

# while Loop

```python
Image('9.png')
```

```python
n = 10
sum = 0
i = 1
while i <= n:
    sum = sum + i
    i = i+1   
print(sum)
```

```python
counter = 0
while counter < 3:
    print("Inside loop")
    counter = counter + 1
else:
    print("Inside else")
```

#### 1. For loop

```python
numbers = [6, 5, 3, 8, 4, 2, 5, 4, 11,65,89,8765,6789,987654,5678,76533456,6543]
sum = 0
for val in numbers:
    sum = sum+val
print("The sum is", sum)
```

#### 2. for loop with break:

```python
Image('11.png')
```

```python
for val in "string":
    if val == "i":
        break
    print(val)
```

#### 3. For loop with continue:

```python
Image('12.png')
```

```python
for val in "string":
    if val == "i":
        continue
    print(val)
```
