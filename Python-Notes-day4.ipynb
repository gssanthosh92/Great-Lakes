#### Function Arguments
You can call a function by using the following types of formal arguments −

        1.Required arguments
        2.Keyword arguments
        3.Default arguments
        4.Variable-length arguments
       

#### 1. Required arguments:

```python
# Function definition is here
def helloeveryone():
    print("Hello everyone")
helloeveryone()
```

#### 2. Keyword arguments:

```python
def greet(nm):
    print("Good ",nm)

greet("Morning")
```

```python
def student(firstname, lastname):  
     print(firstname, lastname)  
    
    
# Keyword arguments                   
student(firstname ='Great', lastname ='Learning')     
student(lastname ='Learning', firstname ='Great')  
```

```python
def func(a, b=5, c=10):
    print('a is', a, 'and b is', b, 'and c is', c)
func(12, 24)
```

```python
def hello(*, name='User'):
    print("Hello %s" % name)
hello(name='Kushal')
```

#### 3. Default arguments

```python
def printinfo( Employeename, Employeeage = 40 ):
   
    print("Employeename: ", Employeename)
    print( "Employeeage ", Employeeage)
    return;


# Now you can call printinfo function
printinfo( "Paul" )
```

```python
# Assign a default value to the function
def def_val_fun(m, n=510): 
    print("m: ", m) 
    print("n: ", n) 
  
#call the function

def_val_fun(110)
```

```python
def test(a , b=-99):
    if a > b:
        return True
    else:
        return False
test(12, 23)
```

#### 4.Variable-length arguments

```python
def printinfo( arg1, *vartuple ):
    print(arg1)
    for var in vartuple:
        print(var)
        return
printinfo( 10 )
printinfo( 70, 60, 50 )   
```

## Scope of variables

In python, the variables are defined with the two types of scopes.

**1. Global variables**

**2. Local variables**

```python
Image('3.png')
```

```python
a = 0 # initialize global var; adds var to global namespace
def local_var():
    a = 3 # initialize local var; adds var to local namespace
    print('local var: a = %s' % a)
local_var()    
```

#### 2. To understand local and global variables we will go through two examples.

```python
def change(b):
    a = 90
    print(a)
a = 9
print("Before the function call ", a)
print("inside change function", end=' ')
change(a)
print("After the function call ", a)
```

```python
def change(b):
    global a
    a = 90
    print(a)
a = 9
print("Before the function call ", a)
print("inside change function", end=' ')
change(a)
print("After the function call ", a)
```

###### global keyword we are telling that a is globally defined, so when we are changing a’s value inside the function it is actually changing for the a outside of the function also.

# lambda function with filter

#### 1.Program to filter out the list which contains odd numbers?

```python
List = {1,2,3,4,10,123,22}  
Oddlist = list(filter(lambda x:(x%3 == 0),List))
print(Oddlist)  
```

#### 2. lambda function to check for the integers that are greater than 7?

```python
numbers_list = [2, 6, 8, 10, 11, 4, 12, 7, 13, 17, 0, 3, 21]

filtered_list = list(filter(lambda num: (num > 7), numbers_list))

print(filtered_list)  
```

## lambda function with map

#### 1.Program to triple each number of the list using map  

```python
List = {1,2,3,4,10,123,22}  
new_list = list(map(lambda x:x*3,List)) 
print(new_list)  
```

#### 2.Create lambda function calculate the remainder after dividing each number by 2?

```python
numbers_list = [2, 6, 8, 10, 11, 4, 12, 7, 13, 17, 0, 3, 21]

mapped_list = list(map(lambda num: num % 2, numbers_list))

print(mapped_list)  
```

```python
list = [10,6,7,5,2,1,8,5]
s = reduce(lambda x,y: x if (x > y) else y, list)
print(s)
```
